```python
import pandas as pd

# Set maximum number of displayed rows
pd.set_option('display.max_rows',50)
```

```python
# Read the CSV file into a DataFrame
df=pd.read_csv(r'C:\Users\ASUS\Desktop\alas imtahan\task_1\vacancy_browser.csv')
df.head()
```

```python
# Display column names
df.columns

df.info()  #general information

df.isna().sum()   # Checking for missing values in each column
```

```python
# Check for duplicated rows
df.duplicated().sum()
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Check for duplicated rows after removal
df.duplicated().sum()

df.info()
```

```python
# Translate Azeri month names to English and convert 'vacancy_date' to datetime format
translate_months = {
    'Yanvar': 'January', 'Fevral': 'February',
    'Mart': 'March', 'Aprel': 'April',
    'May': 'May', 'İyun': 'June',
    'İyul': 'July', 'Avqust': 'August',
    'Sentyabr': 'September', 'Oktyabr': 'October',
    'Noyabr': 'November', 'Dekabr': 'December'
}
df['vacancy_date'] = df['vacancy_date'].replace(translate_months, regex=True)
df['vacancy_date'] = pd.to_datetime(df['vacancy_date'])

df
```

```python
# Extract day, month, and year from 'vacancy_date'
df['day'] = df['vacancy_date'].dt.day
df['month'] = df['vacancy_date'].dt.month_name()
df['year'] = df['vacancy_date'].dt.year
df
```

```python
# Display the top 5 months with the most job vacancies
df['month'].value_counts().head()

# Display the top 10 months and days with the most job vacancies
df[['month', 'day']].value_counts().head(10)
```

```python
# Create a DataFrame containing jobs with 'Qrafik Dizayner' in the job name
df_designer = df[df['vacancy_name'].str.contains('Qrafik .*[Dd]izayner')]

# Create a new column 'salary_without_ISO' to store salaries without currency information
df_designer['salary_without_ISO'] = df_designer['salary'].str.replace('AZN', '').str.replace('minimum', '').str.replace('USD', '')
df_designer
```

```python
# Create a DataFrame without rows where 'salary' column is NaN
df_withoutnan = df.loc[~df['salary'].isna()]
```

```python
# Display rows where 'salary' column does not contain 'AZN'
df_withoutnan[~df_withoutnan['salary'].str.contains('AZN')]

# Display rows where 'salary' column contains 'EUR' or 'USD'
df_withoutnan[df_withoutnan['salary'].str.contains('EUR|USD')]

# Count the number of jobs that offer salaries in EUR or USD
df_withoutnan[df_withoutnan['salary'].str.contains('EUR|USD')].__len__()

# Count the number of jobs that do not offer salaries in EUR or USD
df_withoutnan[~df_withoutnan['salary'].str.contains('EUR|USD')].__len__()
```
